```python
import numpy as np
k=np.log(19999/((20000/345)-1))/50
```

```python
def c(t):
  return 20000 / (1+19999*np.exp(-0.117*t))
```

```python
t=np.arange(0,175,0.01)
```

```python
c(t)
```

```python
dcdt=0.117*c(t)*(1-c(t)/20000)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,12))
plt.plot(t, c(t),c='red')
plt.axvline(np.log(19999)/k,linestyle='--')
plt.axhline(10000,linestyle='--')
plt.title('comportamiento de los contagios', fontsize=16)
plt.xlabel('t',fontsize=16)
plt.ylabel('C',fontsize=16)
```

```python
plt.plot(c(t), dcdt,c='red')
plt.axvline(10000,linestyle='--')
plt.axhline(584,linestyle='--')

```
